2.3 Converting tokens into token IDs

````python
all_words = sorted(set(preprocessed)) # sorted list of unique words
print(all_words[:30])
vocab_size = len(all_words)
print(vocab_size)
````

````python
vocab = {token: integer for integer,token in enumerate(all_words)}
print(vocab)
````

````python
vocab["What"]
````

````python
class SimpleTokenizerV1:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s,i in vocab.items()}
    def encode(self,text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        text = re.sub(r'([,.:;?_!"()\']|--|\s)', r'\1', text)
        return text
````

````python
text = "I am a genius. What are you?"
tokenizer = SimpleTokenizerV1(vocab)
ids = tokenizer.encode(text)
reconstructed = tokenizer.decode(ids)
reconstructed
````

````python
class SimpleTokenizerV2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s,i in vocab.items()}
    def encode(self,text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        preprocessed = [item if item in self.str_to_int else "<|unk|>" for item in preprocessed]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        text = re.sub(r'([,.:;?_!"()\']|--|\s)', r'\1', text)
        return text
````

````python
from torch.utils.data import Dataset, DataLoader

class GPTDatasetV1(Dataset):
    def __init__(self, txt, tokenizer, max_length, stride):
        self.input_ids = []
        self.target_ids = []

        #tokenize the entire text
        token_ids = tokenizer.encode(txt, allowed_special={"<|endoftext|>"})

        #Use a sliding window to chunk the book into overlapping sequences of max_length
        for i in range(0, len(token_ids)-max_length, stride):
            input_chunk = token_ids[i:i+max_length]
            target_chunk = token_ids[i+1:i+1+max_length]
            self.input_ids.append(torch.tensor(input_chunk))
            self.target_ids.append(torch.tensor(target_chunk))
        
    def __len__(self):
        return len(self.input_ids)
    
    def __getitem__(self, idx):
        return self.input_ids[idx], self.target_ids[idx]
    
````

### Linking the parameters to the GPT‑2 tokenizer

In your `create_dataloader_v1`, GPT‑2’s tokenizer is used here:

```python
tokenizer = tiktoken.get_encoding("gpt2")
token_ids = tokenizer.encode(txt, allowed_special={"<|endoftext|>"})
```

So all parameters operate on this **list of GPT‑2 token IDs** (`token_ids`).

---

### Minimal numeric example (with fake GPT‑2 IDs)

Assume GPT‑2 gives you:

```python
token_ids = [10, 20, 30, 40, 50, 60, 70]   # pretend these are GPT-2 tokens
```

Now call:

```python
dataloader = create_dataloader_v1(
    txt=...,         # gives token_ids above
    batch_size=2,
    max_length=4,
    stride=2,
    shuffle=False,
    drop_last=True,
    num_workers=0,
)
```

#### 1. `max_length` – how many GPT‑2 tokens per sequence

Inside `GPTDatasetV1`:

```python
for i in range(0, len(token_ids) - max_length, stride):
    input_chunk  = token_ids[i : i + max_length]
    target_chunk = token_ids[i + 1 : i + max_length + 1]
```

With `max_length = 4` and `stride = 2`:

- `len(token_ids) = 7`, so `range(0, 7 - 4, 2) = range(0, 3, 2) → i = 0, 2`

i = 0:
- `input_chunk  = [10, 20, 30, 40]`  (length 4 → `max_length`)
- `target_chunk = [20, 30, 40, 50]`  (shifted by 1)

i = 2:
- `input_chunk  = [30, 40, 50, 60]`
- `target_chunk = [40, 50, 60, 70]`

So **dataset has 2 items**, each item is:
- inputs:  4 GPT‑2 tokens  
- targets: 4 GPT‑2 tokens (next‑token labels)

**`max_length` = 4** → every sample is a sequence of 4 GPT‑2 tokens.

---

#### 2. `stride` – how far you slide the GPT‑2 window

Above we used `stride = 2`, so we advanced 2 tokens each time (`i = 0, 2, ...`).

If we used `stride = 1` instead:

- `range(0, 7 - 4, 1) = range(0, 3, 1) → i = 0, 1, 2`

You’d get 3 samples (more overlap):

- i = 0: `[10, 20, 30, 40]`
- i = 1: `[20, 30, 40, 50]`
- i = 2: `[30, 40, 50, 60]`

**Smaller stride = more overlapping GPT‑2 sequences, more training examples.**

---

#### 3. `batch_size` – how many sequences per batch

We have 2 dataset items and `batch_size = 2`, so the `DataLoader` collects them into one batch:

```python
inputs  → shape [2, 4]  # 2 sequences, each 4 GPT‑2 tokens
targets → shape [2, 4]
```

If `batch_size = 1`, you’d get 2 batches, each of shape `[1, 4]`.

**So in GPT‑2 terms: one batch = `batch_size` rows of token IDs, each row length = `max_length`.**

---

#### 4. `shuffle`

- `shuffle=False` → dataset order is as created: first window, second window, etc.
- `shuffle=True`  → the order of samples is randomized **before** batching (but each sample is still a GPT‑2 sequence of length `max_length`).

---

#### 5. `drop_last`

Imagine you had 5 samples and `batch_size=2`:

- Batches without `drop_last`:
  - Batch 1: 2 samples
  - Batch 2: 2 samples
  - Batch 3: 1 sample

- With `drop_last=True`, that last 1‑sample batch is **dropped**.

This doesn’t change tokens, just whether the final smaller batch is kept.

---

#### 6. `num_workers`

- `0` → load data in the main process (simplest; fine here).
- `>0` → use background worker processes to prepare batches faster (useful for large datasets).

---

If you’d like, we can plug in your actual `raw_text` and specific values (e.g. `max_length=4`, `stride=1`) and I can walk you through the **exact first 2–3 batches** in terms of GPT‑2 token IDs and decoded text.

````python
from torch.utils.data import Dataset, DataLoader
def create_dataloader_v1(txt, batch_size=4, max_length=256, 
                         stride=128, shuffle=True, drop_last=True,
                         num_workers=0):

    # Initialize the tokenizer
    tokenizer = tiktoken.get_encoding("gpt2")

    # Create dataset
    dataset = GPTDatasetV1(txt, tokenizer, max_length, stride)

    # Create dataloader
    dataloader = DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=shuffle,
        drop_last=drop_last,
        num_workers=num_workers
    )

    return dataloader
````
